```python
batch_size = 64
data_path = './data'
normalize_mean, normalize_std = (0.1307,), (0.3081,)

model_config = './model_config/fc1_cocktail_party_mnist_instance.json'
optimizer_type = 'Adam'
learning_rate = 0.001
criterion_type = 'CrossEntropyLoss'
num_epochs = 20
save_path = './checkpoints/122123_fc1_cocktail_party_mnist_pretraining_wout_bias.pth'
save_interval = 100
linearize = True
```

```python
train_loss_arr, val_loss_arr, val_accuracy_arr = train_all(model_config, optimizer_type, learning_rate, criterion_type, num_epochs, train_loader, val_loader=val_loader, save_path=save_path, save_interval=save_interval, linearize=linearize)
```
